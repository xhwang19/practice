```python
import requests

req = requests.get('https://en.wikipedia.org/wiki/Science')
```

```python
req
```

```python
webpage = req.text
print(webpage)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(webpage, 'html.parser')
```

```python
print(soup.prettify())
```

```python
paragraph = soup.find('p', attrs={'class':False})
paragraph
```

```python
paragraph.find_all('a')
```

```python
paragraph.find_all('a',attrs={'title':True})
```

```python
data = {'title':[], 'href':[]}
for link in paragraph.find_all('a', attrs={'title':True}):
    data['title'].append(link['title'])
    data['href'].append(link['href'])
```

```python
import pandas as pd
df = pd.DataFrame(data)
df
```

```python
df.to_csv('C:/Peter Wang/2023/Machine Learning/Week13/Assignment/Sample_data.csv')
```

```python
webpages = []
head = 'https://en.wikipedia.org'
for href in data['href']:
    link = head + href
    req = requests.get(link)
    webpage = req.text
    webpages.append(webpage)
```
